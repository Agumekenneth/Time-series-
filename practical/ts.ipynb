 # Converting Date to Datetime and Setting Time Index

# Transforming the data and stationarity

```python
# Selecting one stock (AAPL)
df_aapl = df[df['Ticker'] == 'AAPL']
```

```python
# Keep only the Close price
ts = df_aapl['Close']
```

```python
# View first few values
ts.head()
```

- Plotting the raw time series

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(ts)
plt.title("AAPL Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

- Checking stationarity using ADF Test

```python
from statsmodels.tsa.stattools import adfuller

# Performming ADF test
adf_result = adfuller(ts)
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
```

- This shows the data of  time series is non-stationary

- To fix non-stationarity, we apply first- order differencing.

```python
# Apply first differencing
ts_diff = ts.diff()

# Remove the first NaN value
ts_diff = ts_diff.dropna()

# Plot differenced series
plt.figure(figsize=(10, 4))
plt.plot(ts_diff)
plt.title("Differenced AAPL Close Price")
plt.xlabel("Date")
plt.ylabel("Differenced Value")
plt.show()
```

 - Re-checking stationarity after differencing

```python
# ADF test after differencing
adf_result_diff = adfuller(ts_diff)

print("ADF Statistic (After Differencing):", adf_result_diff[0])
print("p-value (After Differencing):", adf_result_diff[1])
```

# Line Plot 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(ts, label='Close Price')
plt.title("Line Plot of AAPL Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```

- The line plot provides an overall view of the time series. It helps identify trends, fluctuations, and unusual values. The plot shows that the stock price changes over time with noticeable trends, indicating non-stationary behavior

# Seasonal Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(ts, model='additive', period=30)
decomposition.plot()
plt.show()
```

- Seasonal decomposition was applied to break the time series into trend, seasonal, and residual components. The trend component shows long-term movement, while the seasonal component captures repeating patterns. The residual represents random fluctuations

# Autocorrelation (ACF) & Partial Autocorrelation (PACF)

- This is used  to understand lag relationships

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.figure(figsize=(12, 4))
plot_acf(ts, lags=40)
plt.show()

plt.figure(figsize=(12, 4))
plot_pacf(ts, lags=40, method='ywm')
plt.show()
```

- Slow decay in ACF  non-stationary
- Significant spikes  strong lag dependency

- The ACF and PACF plots were used to examine the correlation between current values and past observations. Significant spikes at early lags indicate strong autocorrelation, which is common in financial time series data.

# Seasonal Heatmap

- This is used to show patterns across months and years 

```python
import seaborn as sns

# Create a copy for heatmap
heat_df = df_aapl.copy()
heat_df['Year'] = heat_df.index.year
heat_df['Month'] = heat_df.index.month_name()

pivot = heat_df.pivot_table(
    values='Close',
    index='Year',
    columns='Month',
    aggfunc='mean'
)

# Correct month order
months_order = ['January','February','March','April','May','June',
                'July','August','September','October','November','December']
pivot = pivot.reindex(columns=months_order)

plt.figure(figsize=(14, 6))
sns.heatmap(pivot, cmap='coolwarm')
plt.title("Seasonal Heatmap of AAPL Close Price")
plt.show()
```

- A seasonal heatmap was used to visualize average stock prices across months and years. This visualization helps identify periods of high and low price intensity and reveals seasonal patterns

# Distribution Plots (Histogram and Box Plot)

- This is used to check normality,skewness and outliers

```python
plt.figure(figsize=(12, 4))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(ts, kde=True)
plt.title("Histogram of Close Price")

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=ts)
plt.title("Box Plot of Close Price")
plt.tight_layout()
plt.show()
```

- The histogram and box plot were used to analyze the distribution of the time series. The histogram shows the distribution shape, while the box plot helps identify potential outliers in the data.
