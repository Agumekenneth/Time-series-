# Checking for outliers

```python
cont_data = df.select_dtypes(include=['float64', 'int64'])
cat_data = df.select_dtypes(include=['object'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
for i, col in enumerate(cont_data.columns):
    plt.subplot(5,4, i+1)
    sns.boxplot(cont_data[col])
    plt.title(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
def remove_outliers(columns):
    for column in columns:
        Q1 = cont_data[column].quantile(0.25)
        Q3 = cont_data[column].quantile(0.75)

        IQR = Q3 - Q1
        upper_bound = Q3 + 1.5 * IQR
        lower_bound = Q1 - 1.5 * IQR

        cont_data[column] = np.where(
            cont_data[column] > upper_bound, upper_bound, cont_data[column]
        )
        cont_data[column] = np.where(
            cont_data[column] < lower_bound, lower_bound, cont_data[column]
        )
```

```python
remove_outliers(cont_data)
```

```python
plt.figure(figsize=(12, 12))
for i, col in enumerate(cont_data.columns):
    plt.subplot(5, 4, i+1)
    plt.boxplot(cont_data[col])
    plt.title(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

- Outliers were analyzed using boxplots and the Interquartile Range (IQR) method on continuous variables. Instead of removing outliers, extreme values were capped using upper and lower IQR bounds (Winsorization). This approach preserves the temporal structure of the time series while reducing the influence of extreme values, which is particularly important for financial data where abrupt price changes may reflect real market events

# Transforming the data and stationarity

```python
# Selecting one stock (AAPL)
df_aapl = df[df['Ticker'] == 'AAPL']
```

```python
# Keep only the Close price
ts = df_aapl['Close']
```

```python
# View first few values
ts.head()
```

- Plotting the raw time series

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(ts)
plt.title("AAPL Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

- Checking stationarity using ADF Test

```python
from statsmodels.tsa.stattools import adfuller

# Performming ADF test
adf_result = adfuller(ts)
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
```

- This shows the data of  time series is non-stationary

- To fix non-stationarity, we apply first- order differencing.

```python
# Apply first differencing
ts_diff = ts.diff()

# Remove the first NaN value
ts_diff = ts_diff.dropna()

# Plot differenced series
plt.figure(figsize=(10, 4))
plt.plot(ts_diff)
plt.title("Differenced AAPL Close Price")
plt.xlabel("Date")
plt.ylabel("Differenced Value")
plt.show()
```

 - Re-checking stationarity after differencing

```python
# ADF test after differencing
adf_result_diff = adfuller(ts_diff)

print("ADF Statistic (After Differencing):", adf_result_diff[0])
print("p-value (After Differencing):", adf_result_diff[1])
```

# Line Plot 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(ts, label='Close Price')
plt.title("Line Plot of AAPL Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```

- The line plot acts as a heartbeat check of the time series. It reveals long-term trends, volatility, and sudden price changes. In this dataset, the plot shows clear fluctuations and upward and downward movements, which is typical of financial time series data

# Seasonal Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(ts, model='additive', period=30)
decomposition.plot()
plt.show()
```

- The decomposition shows that while the trend captures the long-term movement of stock prices, the seasonal component is relatively weak, indicating limited fixed periodic behavior. The residual component contains irregular movements caused by market shocks and investor behavior

# Autocorrelation (ACF) & Partial Autocorrelation (PACF)

- This is used  to understand lag relationships

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.figure(figsize=(12, 4))
plot_acf(ts, lags=40)
plt.show()

plt.figure(figsize=(12, 4))
plot_pacf(ts, lags=40, method='ywm')
plt.show()
```

- Slow decay in ACF  non-stationary
- Significant spikes  strong lag dependency

- The ACF plot shows a slow decay, which confirms the presence of trend and non-stationarity in the original series. The PACF plot shows significant spikes at lower lags, indicating direct dependence on recent past values. This behavior is typical in financial time series and justifies the need for differencing

# Seasonal Heatmap

- This is used to show patterns across months and years 

```python
import seaborn as sns

# Create a copy for heatmap
heat_df = df_aapl.copy()
heat_df['Year'] = heat_df.index.year
heat_df['Month'] = heat_df.index.month_name()

pivot = heat_df.pivot_table(
    values='Close',
    index='Year',
    columns='Month',
    aggfunc='mean'
)

# Correct month order
months_order = ['January','February','March','April','May','June',
                'July','August','September','October','November','December']
pivot = pivot.reindex(columns=months_order)

plt.figure(figsize=(14, 6))
sns.heatmap(pivot, cmap='coolwarm')
plt.title("Seasonal Heatmap of AAPL Close Price")
plt.show()
```

The seasonal heatmap compares average prices across months and years. Variations in color intensity indicate periods of higher and lower average prices, allowing easy identification of temporal patterns and market cycles
