```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#Loading the dataset
df = pd.read_csv('TS_Data_Q1.csv')
df
```

```python
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
```

```python
df
```

```python
# Plotting and describing the time series
plt.figure(figsize=(10,5))
plt.plot(df['Passengers'], marker='o')
plt.title("Monthly Airline Passengers")
plt.xlabel("Date")
plt.ylabel("Number of Passengers")
plt.show()

print("Basic Description:")
print(df['Passengers'].describe())
```

- The graph shows airline passengers increasing dramatically from 1970.
- A clear seasonal pattern with peaks and troughs each year.
- The seasonal fluctuations get larger as passenger numbers increases.

```python
# Classical decomposition
decomposition = seasonal_decompose(df['Passengers'], model='mu', period=12)

# Plot all components in a single figure
fig = decomposition.plot()
fig.set_size_inches(14, 8)
plt.show()
```

- The trend shows passengers growing steadily from around 120 to 200 over this period.
- The seasonal component repeats every 12 months ie summer months especially  get a boost of +15 to +18 passengers while winter months decreases around -5 to -12.
- Therefore this shows summers are always busy, winters are quieter and there's barely any random noise left over
