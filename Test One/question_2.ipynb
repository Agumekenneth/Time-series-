```python
# Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Load the dataset
df = pd.read_csv("TS_Data_Q2.csv")
df
```

```python
# selecting 'Price' for closing prices
ts = df['Price']
```

```python
#ACF of raw data
plt.figure(figsize=(10,4))
plot_acf(ts, lags=40)
plt.title("ACF of Raw Price Data")
plt.show()
```

- The ACF of the raw data decays very slowly and stays significant for many lags thus indicating that the series has a trend and isn’t stationary

```python
#First-order differencing
ts_diff = ts.diff().dropna()  # removing NaN after differencing
plt.figure(figsize=(10,4))
plot_acf(ts_diff, lags=40)
plt.title("ACF of Differenced Series")
plt.show()

plt.figure(figsize=(10,4))
plot_pacf(ts_diff, lags=40)
plt.title("PACF of Differenced Series")
plt.show()
```

- After first-order differencing, both the ACF and PACF plots show a significant spike only at lag 1, with no significant correlations at higher lags.
- This suggests that the differenced series may be white noise thus indicating that the original series follows a random walk
